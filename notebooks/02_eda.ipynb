```python
import pandas as pd
```

```python
df = pd.read_csv('../data/cleaned_churn_data.csv')
df.head()
```

```python
df['churn'].value_counts()
```

```python
df['churn'].value_counts(normalize=True) * 100

```

```python
df.groupby('contract_type')['churn'].mean() * 100
#“Customers with month-to-month contracts show the highest churn.”
```

```python
df.groupby(
    pd.cut(df['tenure'], bins=[0,6,12,24,60]),
    observed=False
)['churn'].mean() * 100
```

```python
df.groupby(pd.cut(df['monthly_charges'], bins=5))['churn'].mean() * 100
```
