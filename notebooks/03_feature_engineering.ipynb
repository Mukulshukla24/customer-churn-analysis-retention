```python
import pandas as pd
```

```python
df = pd.read_csv('../data/cleaned_churn_data.csv')
df.head()
```

```python
df['tenure_group'] = pd.cut(
    df['tenure'],
    bins=[0, 6, 12, 24, 60],
    labels=['0-6', '6-12', '12-24', '24+']
)
```

```python
df[['tenure', 'tenure_group']].head(10)
```

```python
median_charge = df['monthly_charges'].median()
median_charge
```

```python
df['high_monthly_charge'] = (df['monthly_charges'] > median_charge).astype(int)
```

```python
df[['monthly_charges', 'high_monthly_charge']].head()
```

```python
service_cols = ['internet_service', 'streaming', 'security', 'tech_support']
```

```python
df[service_cols] = df[service_cols].apply(lambda x: x.map({'Yes': 1, 'No': 0}))
df['service_count'] = df[service_cols].sum(axis=1)
```

```python
df[['internet_service', 'streaming', 'security', 'tech_support', 'service_count']].head()
```

```python
df['is_monthly_contract'] = (df['contract_type'] == 'Month-to-month').astype(int)
```

```python
df[['contract_type', 'is_monthly_contract']].head()
```

```python
df.head()
```

```python
df.to_csv('../data/feature_ready_churn_data.csv', index=False)
```
