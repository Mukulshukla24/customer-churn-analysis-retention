```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "postgresql://postgres:#Mukul2409@localhost:5432/churn_analysis"
)
```

```python
query = """
SELECT
    c.customer_id,
    c.gender,
    c.senior_citizen,
    c.tenure,
    c.contract_type,
    b.monthly_charges,
    b.total_charges,
    s.internet_service,
    s.streaming,
    s.security,
    s.tech_support,
    st.churn
FROM customers c
JOIN billing b ON c.customer_id = b.customer_id
JOIN services s ON c.customer_id = s.customer_id
JOIN status st ON c.customer_id = st.customer_id
"""
df = pd.read_sql(query, engine)
df.head()
```

```python
df['total_charges'] = pd.to_numeric(df['total_charges'], errors='coerce')
```
